```python
import numpy as np
import matplotlib.pyplot as plt

methods = [
    "Iterative LDA Method",
    "Iterative LSA Method",
    "GA LDA Method [Config 5]",
    "GA LsA Method [5 Topics]",
    "Topic Wise LDA Method [11 Topics]",
    "Topic Wise LSA Method [12 Topics]",
    "Gird Search LDA Method",
    "Gird Search LSA Method [5 Topics]"
]

classifiers = [
    ["LinearSVC + 3 Fold Cross Validation", "LinearSVC + MultinominalNB"],
    ["LinearSVC + RandomForest"],
    ["LinearSVC + RandomForest + 3 Fold Cross Validation"],
    ["LinearSVC + 3 Fold Cross Validation"],
    ["LinearSVC + 3 Fold Cross Validation"],
    ["LinearSVC + 3 Fold Cross Validation"],
    ["LinearSVC + RandomForest"],
    ["LinearSVC + 3 Fold Cross Validation"]
]

accuracies = [
    [0.7759901882981026, None],
    [0.7960103888608326],
    [0.8532],
    [0.852, None],
    [0.852, None],
    [0.868, None],
    [0.8167972224861167],
    [0.852]
]

# Assigning unique colors to each classifier
color_dict = {
    "LinearSVC + 3 Fold Cross Validation": 'tab:blue',
    "LinearSVC + MultinominalNB": 'tab:orange',
    "LinearSVC + RandomForest": 'tab:green',
    "LinearSVC + RandomForest + 3 Fold Cross Validation": 'tab:red'
}

# Assigning unique marker styles to each method
marker_dict = {
    "Iterative LDA Method": 'o',
    "Iterative LSA Method": '^',
    "GA LDA Method [Config 5]": 's',
    "GA LsA Method [5 Topics]": 'd',
    "Topic Wise LDA Method [11 Topics]": 'p',
    "Topic Wise LSA Method [12 Topics]": '*',
    "Gird Search LDA Method": 'x',
    "Gird Search LSA Method [5 Topics]": 'h'
}

# Create radar chart
plt.figure(figsize=(10, 10))
plt.title('Performance Comparison of Methods by Classifiers')

for method, classifier_list, accuracy_list in zip(methods, classifiers, accuracies):
    for classifier, accuracy in zip(classifier_list, accuracy_list):
        if accuracy is not None:
            plt.scatter(method, accuracy, color=color_dict[classifier], marker=marker_dict[method], label=classifier)

# Custom legend for classifiers
handles = []
labels = []
for classifier, color in color_dict.items():
    handles.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10))
    labels.append(classifier)

plt.legend(handles, labels, loc='upper left')

plt.xlabel('Methods')
plt.ylabel('Accuracy')

plt.xticks(rotation=45, ha='right')

plt.grid(True)
plt.tight_layout()
plt.show()
```
